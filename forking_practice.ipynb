## Step 3 â€” A Little Python While You're Here

You've verified your fork. Here's a quick function that summarizes what you just did.

Read the code, then run it.

````python
def describe_fork_workflow(original_owner: str, repo_name: str, your_username: str) -> None:
    """
    Print a description of the fork â†’ clone â†’ push workflow.

    Parameters
    ----------
    original_owner : str
        GitHub username of the person who owns the original repo.
    repo_name : str
        Name of the repository.
    your_username : str
        Your GitHub username.
    """
    original = f"github.com/{original_owner}/{repo_name}"
    your_fork = f"github.com/{your_username}/{repo_name}"
    your_local = f"your laptop/{repo_name}/"

    print("The Fork Workflow")
    print("=" * 50)
    print(f"  Original repo  â†’  {original}")
    print(f"  Your fork      â†’  {your_fork}")
    print(f"  Local copy     â†’  {your_local}")
    print()
    print("Your changes stay in YOUR fork.")
    print("The original is untouched â€” that's the whole point!")


# Call the function using your info from Step 1
describe_fork_workflow(
    original_owner="instructor",
    repo_name="forking-practice",
    your_username=your_github_username,
)
````
